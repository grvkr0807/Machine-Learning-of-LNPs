```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import skew, kurtosis, shapiro, kstest, anderson, norm, probplot


file_path = r'C:\Users\grvkr\Box\Gaurav Kumar\Purdue_Work\SAR_NM\Data\Anderson_group_Data5_2022\Activity\InVitro_Activity.xlsx'  # Input file path
sheet_name = "Z-score Normalized"  # Sheet containing normalized data
data_raw = pd.read_excel(file_path, sheet_name=sheet_name, engine='openpyxl')
data = data_raw.values.flatten()  # Convert from 2D to 1D


# Calculate skewness and kurtosis
skewness = skew(data)
kurt = kurtosis(data)  # By default, this is excess kurtosis (subtract 3 for standard kurtosis)

# Perform Shapiro-Wilk Test
stat_shapiro, p_shapiro = shapiro(data)

# Perform Kolmogorov-Smirnov Test
stat_ks, p_ks = kstest(data, 'norm', args=(np.mean(data), np.std(data)))

# Perform Anderson-Darling Test
result_ad = anderson(data)

# Print results
print(f"Skewness: {skewness}")
print(f"Kurtosis: {kurt}")
print(f"Shapiro-Wilk Test: statistic={stat_shapiro}, p-value={p_shapiro}")
print(f"Kolmogorov-Smirnov Test: statistic={stat_ks}, p-value={p_ks}")
print(f"Anderson-Darling Test: statistic={result_ad.statistic}, critical values={result_ad.critical_values}")

# Visual inspection: Histogram and Q-Q plot
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Histogram
axes[0].hist(data, bins=30, density=True, alpha=0.75, color='skyblue', edgecolor='black')
axes[0].set_title('Histogram with Density')

# Q-Q plot
probplot(data, dist="norm", plot=axes[1])
axes[1].set_title('Q-Q Plot')

plt.tight_layout()
plt.show()

# Assess normality function
def assess_normality(skewness, kurt, p_shapiro, p_ks, ad_stat, ad_crit):
    normal = True
    
    if -0.5 < skewness < 0.5 and -1.0 < kurt < 1.0:
        print("Skewness and kurtosis suggest normal distribution.")
    else:
        print("Skewness and kurtosis suggest non-normal distribution.")
        normal = False
    
    if p_shapiro > 0.05:
        print("Shapiro-Wilk Test suggests normal distribution.")
    else:
        print("Shapiro-Wilk Test suggests non-normal distribution.")
        normal = False
    
    if p_ks > 0.05:
        print("Kolmogorov-Smirnov Test suggests normal distribution.")
    else:
        print("Kolmogorov-Smirnov Test suggests non-normal distribution.")
        normal = False
    
    if ad_stat < ad_crit[2]:  # 5% significance level
        print("Anderson-Darling Test suggests normal distribution.")
    else:
        print("Anderson-Darling Test suggests non-normal distribution.")
        normal = False
    
    return normal

normality = assess_normality(skewness, kurt, p_shapiro, p_ks, result_ad.statistic, result_ad.critical_values)
print(f"Overall assessment: {'Normal distribution' if normality else 'Non-normal distribution'}")



```
